```python
# Import pandas and numpy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Loading the dataset
data = pd.read_csv('autolib.csv')

# Creating a dataframe

dataf = pd.DataFrame(data)
```

```python
# Preview the data
dataf.head(7)
```

DATA PREPARATION

```python
#Renaming the columns 
#1. Concatenate the year,month and day to be on one column with the datetime object.
dataf['Date']=dataf.apply(lambda x:'%s/%s/%s' % (x['year'],x['month'],x['day']),axis=1)


# Function to confirm dates ate okay
def confirm_date(dataf,Date):
  dataf[Date] = dataf[Date].apply(lambda x :
                                  dataf.datetime.strif(x,'%y/%m/%d'))
dataf.head()

```

```python
#Combining hour and minute to be one column time
dataf['Time'] = dataf.apply(lambda a : '%s:%s' % (a['hour'], a['minute']), axis = 1)
dataf.head()

def confirm_date(dataf,Date):
  dataf[Date] = dataf[Date].apply(lambda x :
                                  dataf.datetime.strif(x,'%H : % M'))
dataf.head()

```

DATA CLEANING

Validity - Dealing with the irrelevant data

```python
# Dropping unnecessary columns
#1. Dropping the geo points column since it will have no effect on the study. We have public name
dataf.drop(columns = 'Geo point', inplace = True)
dataf.head()
```

```python
#Dropping the year,month,day , hour and minute column since we have the date time column
dataf.drop(columns = ['year','month', 'day', 'hour', 'minute'], inplace = True)
dataf.head()
```

COMPLETENESS
Checking and handling missing values

```python
#Checking for missing values
def missing(dataf):
  missingu = dataf.isnull.sum()/len(dataf).sort_values(ascending = True)
print(missing)

```

```python
#Replacing missing values with 0
dataf['Scheduled at'].fillna(0, inplace = True)
dataf['Displayed comment'].fillna(0, inplace = True)
dataf.head()
```
